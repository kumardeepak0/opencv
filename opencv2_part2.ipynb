```python
import numpy as np
```

```python
# Translation matrix the first row moves pixels horizontally and the second row moves in vertically.
M=np.float32([[1,0,100],[0,1,150]])
M
```

```python
shifted=cv2.warpAffine(image, M , (image.shape[1],image.shape[0]))
plt.imshow(shifted)
```

```python
#Changed the pixel directions to move image left and up
#[1, 0, -50] [row,column,number of pixels]
M=np.float32([[1,0,-150],[0,1,-100]])
shifted=cv2.warpAffine(image, M , (image.shape[1],image.shape[0]))
plt.imshow(shifted)
plt.show()
```

```python
(height,width,channels)=image.shape
center=(width//2,height//2)
center
```

```python
M=cv2.getRotationMatrix2D(center,45,1.0)
rotated=cv2.warpAffine(image,M,(width,height))
plt.imshow(rotated)
plt.show()
```

```python
M=cv2.getRotationMatrix2D(center,60,2.0)
rotated=cv2.warpAffine(image,M,(width,height))
plt.imshow(rotated)
plt.show()
```

```python
M=cv2.getRotationMatrix2D(center,180,1.0)
rotated=cv2.warpAffine(image,M,(width,height))
plt.imshow(rotated)
```

```python
M = cv2.getRotationMatrix2D(center, -90, 1.0)
rotated = cv2.warpAffine(image, M, (width, height))
plt.imshow(rotated)
```

**Resizing**

```python
#Here we rezise setting width to 150 and scale the height accordingly
```

```python
r=150.0/ image.shape[1]
dim=(150,int(image.shape[0]*r))
resized=cv2.resize(image,dim,interpolation=cv2.INTER_AREA)
plt.imshow(resized)
plt.show()
```

```python
#Here we rezise setting height to 150 and scale the width accordingly
```

```python
r = 150.0 / image.shape[0]
dim = ( int(image.shape[1] * r), 150)
resized = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
plt.imshow(resized)
plt.show()
```

**FLIPPING**

```python
# Flipping across y-axis
flipped = cv2.flip(image,1)
plt.imshow(flipped)
plt.show()
```

```python
# Flipping along X-axis
flipped=cv2.flip(image,0)
plt.imshow(flipped)
plt.show()
```

```python
# flipping along X and y axis
flipped=cv2.flip(image,-1)
plt.imshow(flipped)
plt.show()
```
